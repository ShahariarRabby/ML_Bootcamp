```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
z = x ** 3
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('X Label')
plt.ylabel('Y lable')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,z)
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('X label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,.8,.8])
axes2 = fig.add_axes([.2,0.4,.4,.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
for i in axes[1]:
    i.plot(x,y)
```

```python
ax = plt.figure()
axs = ax.add_axes([.1,.1,.8,.8])
axs.plot(x,y,label='SQR')
axs.plot(x,z,label='Cube')
axs.plot(x,x,label='none')

axs.legend(loc=0)
```
